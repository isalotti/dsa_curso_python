### Gráfico de linha

```python
x = linspace(0, 5, 10)
y = x ** 2

# Cria a figura
fig = plt.figure()

# Define a escala dos eixos
axes = fig.add_axes([0,0,0.8,0.8])

# Cria o plot
axes.plot(x, y, 'r')

# Labels e título
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Gráfico de Linha')
```

```python
# Gráficos de linha com 2 figuras

x = linspace(0,5,10)
y = x ** 2

# Cria a figura
fig = plt.figure()

#Eixos
axes1 = fig.add_axes([0.1, 0.2, 0.8, 0.8]) #figura principal
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) #figura secundária

# Fig principal
axes1.plot(x, y, 'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes.set_title('Figura Principal')

# Fig secundária
axes2.plot(y,x,'g')
axes2.set_xlabel('y')
axes2.set_ylabel('x')
axes2.set_title('Figura Secundária')
```

```python
# Gráficos de linha em paralelo

x = linspace(0,5,10)
y = x ** 2

# Divide a area de plotagem em 2
fig , axes = plt.subplots(nrows = 1, ncols = 2)

# Loop pelos eixos para criar cada plot
for ax in axes:
    ax.plot(x, y, 'r')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('Título')
    
# Ajustar layout
fig.tight_layout()
```

```python
# Gráficos de linha com diferentes escalas

x = linspace(0, 5, 10)
y = x ** 2

fig, axes = plt.subplots(1, 2, figsize = (10,4))
# 1 , nlines
# 2 = ncols

# Plot1
axes[0].plot(x, x**2, x, exp(x))
axes[0].set_title("Escala Padrão")

#Plot 2
axes[1].plot(x, x**2, x, exp(x))
axes[1].set_yscale("log")  #Escala na base de logarítimo
axes[1].set_title("Escala Locaritmica (y)")
```

```python
# Grid

x = linspace(0, 5, 10)
y = x ** 2

# subplots
fig, axes = plt.subplots(1, 2, figsize = (10,3))

#Grid padrão
axes[0].plot(x, x **2, x, x**3, lw = 2)
axes[0].grid(True)

#Grid custom
axes[1].plot(x, x**2, x, x**3, lw = 2)
axes[1].grid(color = 'b', alpha = 0.7, linestyle = 'dashed', linewidth = 0.8)
```

```python
# Diferentes estilos de plot

xx = np.linspace(-0.75, 1., 100)
n = np.array([0,1,2,3,4,5])

#Subplots
fig, axes = plt.subplots(1, 4, figsize = (12,3))

# Plot 1
axes[0].scatter(xx, xx + 0.25 * randn(len(xx)), color = "black")
axes[0].set_title("scatter")

# Plot 2
axes[1].step(n, n**2, lw = 2, color = "blue")
axes[1].set_title("step")

# Plot 3
axes[2].bar(n, n**2, align = "center", width = 0.5, alpha = 0.5, color = "magenta")
axes[2].set_title("bar")

# Plot 4
axes[3].fill_between(x, x**2, x**3, alpha = 0.5, color = "green")
axes[3].set_title("fill_between")  # preenche entre as linhas
```

## Histogramas


```python
# Histogramas

# Dados
n = np.random.randn(1000000)

# Cria os subplots
fig , axes = plt.subplots(1, 2, figsize = (12,4))

# Plot 1
axes[0].hist(n)
axes[0].set_title("Histograma Padrão")
axes[0].set_xlim((min(n), max(n)))     # limites mínimos e máximos para o eixo x

# Plot 2
axes[1].hist(n, cumulative = True, bins = 50)   #Gráfico acumulado, e caixinhas (bins) = 50
axes[1].set_title("Histograma Cumulativo")
axes[1].set_xlim((min(n), max(n)))
```

## Gráficos em 3D 

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

```python
# Dados
alpha = 0.7
phi_ext = 2 * np.pi * 0.5

# Função para um mapa de cores
def ColorMap(phi_m, phi_p):
    return (+ alpha - 2 * np.cos(phi_p) * cos(phi_m) - alpha * np.cos(phi_ext - 2*phi_p))

# Outros dados
phi_m = np.linspace(0,2*np.pi,100)
phi_p = np.linspace(0,2*np.pi,100)
X, Y = np.meshgrid(phi_p, phi_m)   #juntando os dados, para colocar no gráfico
Z = ColorMap(X, Y).T      # a dim Z é a geração das cores
```

```python
# Cria figura
fig = plt.figure(figsize = (14,6))

# Adiciona subplot 1 com projeção 3d
ax = fig.add_subplot(1, 2, 1, projection = '3d')
p = ax.plot_surface(X, Y, Z, rstride = 4, cstride = 4, linewidth = 0)

# Adiciona subplot 2 com projeção 3d
ax = fig.add_subplot(1, 2, 2, projection = '3d')
p = ax.plot_surface(X, Y, Z, rstride = 1, cstride = 1, cmap=cm.coolwarm, linewidth = 0, antialiased=False)

# Cria a barra de cores como legenda
cb = fig.colorbar(p, shrink=0.5)      #Legenda do segundo plot
```

```python
# Wire frame  - sem preenchimento no meio
fig = plt.figure(figsize=(8,6))

# Subplot 
ax = fig.add_subplot(1, 1, 1, projection = '3d')

# Wire frame
p = ax.plot_wireframe(X, Y, Z, rstride=4, cstride=4)
```
