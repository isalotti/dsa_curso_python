# Seaborn

```python
# Importando a lib do Seaborn
import seaborn as sea
sea.__version__
```

```python
# Importando o Pandas
import pandas as pd
pd.__version__
```

```python
# Carregar um dataset que vem com o Seaborn
dados = sea.load_dataset("tips")
```

```python
dados.head()
```

```python
# Mtodo joinplot cria plot de 2 variveis com grficos bivariados e univariados
sea.jointplot(data = dados, x = "total_bill", y = "tip", kind='reg')
```

```python
# Mtodo lmplot() cria plot c/ dados e modelos de regresso
sea.lmplot(data = dados, x = "total_bill", y = "tip", col = "smoker")  # A relao das 2 em comparao com smoker (col)
```

```python
# Dataframe com Pandas
df = pd.DataFrame()
```

```python
# Alimentando o DataFrame com valores aleatrios
import random
df['idade'] = random.sample(range(20,100),30)
df['peso'] = random.sample(range(55,150),30)
```

```python
df.shape
```

```python
df.head()
```

```python
# lmplot
sea.lmplot(data = df, x = "idade", y = "peso", fit_reg = True)  #fit_reg = cria modelo de regresso
```

```python
# kdeplot  - Grfico de densidade
sea.kdeplot(df.idade)
```

```python
# kdeplot
sea.kdeplot(df.peso)
```

```python
# distplot - une histograma e grfico de densidade
sea.distplot(df.peso)    #distplot() -> deprecated
```

```python
# Mtodo distplot()  deprecated.
# Soluo usando histplot:
# Fonte: https://stackoverflow.com/questions/67638590/emulating-deprecated-seaborn-distplots

sea.histplot(df.peso, alpha=0.4, kde=True, kde_kws={"cut":2}, linewidth=0)

# kde = insere um plot de densidade (a linha)
# kde_kws = para deixar uma margem antes e depois no grfifico de histograma, seno fica tudo ajustado nas margens
# linewidth = retirar / colocar linha de separao entre as colunas do histograma
```

```python
# Histograma

# Coleo de funes e estilos do Matplotlib
import matplotlib.pyplot as plt
# % (operador mgico do jupyter notebook) vai criar os grficos nesta pgina do jupyter notebook  
%matplotlib inline  

plt.hist(df.idade, alpha = .3)  #fazendo o histograma com o Matplotlib

# Com SeaBorn
sea.rugplot(df.idade)  # tracinhos no histograma, indicando os valores achados
```

```python
# Boxplot
sea.boxplot(df.idade, color = 'm')
```

```python
# Box plot
sea.boxplot(df.peso, color = "b")
```

```python
# Violon plot
sea.violinplot(df.idade, color = 'c')
```

```python
# Violin plot
sea.violinplot(df.peso, color = '#DA9FE9')
```

```python
# Clustermap  - visualizar hierarquias
sea.clustermap(df)
```

## Usando Matplotlib, Seaborn, NumPy e Pandas na criao de grfico estatstico

```python
import seaborn as sea
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
#Para exibir o grfico dentro do notebook!
```

```python
# Valores randmicos

np.random.seed(42)   # seed = semente para randomizar
n = 1000
pct_smokers = 0.2   # 0.2% sejam de fumantes

# Variveis
flag_fumante = np.random.rand(n) < pct_smokers   # aqui vai ser True / False
idade = np.random.normal(40, 10, n)   # valor min, mx , conjunto de dados para randomizar
altura = np.random.normal(170,10, n)
peso = np.random.normal(70,10,n)

# Dataframe
dados = pd.DataFrame({'altura': altura, 'peso': peso, 'flag_fumante': flag_fumante}) #dicionrio

# Cria os dados para a varivel flag_fumante
dados['flag_fumante'] = dados['flag_fumante'].map({True:'Fumante', False: 'No Fumante'}) #Converte p/ labels
```

```python
dados.shape
```

```python
dados.head()
```

```python
# Style  - seaborn
sea.set(style = "ticks")

#lmplot  - seaborn (plotar o grfico)
sea.lmplot(x = 'altura',
           y = 'peso',
           data = dados,
           hue = 'flag_fumante',
           palette = ['tab:blue', 'tab:orange'],
          height = 7)
# hue -> com base em qual varivel vai colorir (categoria)

# Labels e ttulo  - matplotlib
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')
plt.title('Relao Entre Altura e Peso de Fumantes e No Fumantes')

# Remove as bordas
sea.despine()

plt.show()   #matplotlib

# Por qu preciso do matplotlib para mostrar o grfico em tela?
# R. Plots created using seaborn need to be displayed like ordinary matplotlib plots. 
# Fonte: https://stackoverflow.com/questions/26597116/seaborn-plots-not-showing-up
```
