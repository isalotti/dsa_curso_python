## Pergunta de Negócio 1:

### Qual Cidade com Maior Valor de Venda de Produtos da Categoria 'Office Supplies'?

```python
query1 = df_dsa[df_dsa['Categoria'] == 'Office Supplies'] \
         .groupby('Cidade')['Valor_Venda'] \
         .sum()
```

```python
# Converte o objeto Series da query em DataFrame novamente
query1 = query1.to_frame()  
# Ordena por maior Valor_Venda acumulado, de forma Descendente
query1 = query1.sort_values(by=['Valor_Venda'],ascending=False)
# Obtém a primeira linha apenas - maior valor
query1.iloc[:1] 
```

```python
## Solução do professor

# Filtrar o Dataframe com registro da categoria desejada:
df_dsa_p1 = df_dsa[df_dsa['Categoria'] == 'Office Supplies']
# Agrupar por cidade e calcular total de valor de venda
df_dsa_p1_total = df_dsa_p1.groupby('Cidade')['Valor_Venda'].sum()
# Encontrar a cidade com maior valor de venda
cidade_maior_venda = df_dsa_p1_total.idxmax()
print("Cidade com maior valor de venda para 'Office Suplies': ", cidade_maior_venda)
```

```python
# Conferindo o resultado:
df_dsa_p1_total.sort_values(ascending = False)
```

## Pergunta de Negócio 5:

### Qual Segmento Teve o Maior Total de Vendas?

Demonstre o resultado através de um gráfico de pizza.

```python
query5 = df_dsa.groupby('Segmento')['Valor_Venda'] \
                .sum() \
                .reset_index()
query5 = query5.sort_values(by=['Valor_Venda'],ascending=False)

# Plot
fig , ax = plt.subplots(figsize=(16,6))
colors =['#90EE90', '#DDA0DD','#ADD8E6']
ax.pie(x = query5['Valor_Venda'], labels=query5['Segmento'], 
       autopct='%1.1f%%', 
       shadow=True, 
       startangle=80, 
       colors=colors)
plt.title('Total de Vendas por Segmento')
plt.show()
```

```python
## Solução do professor

df_dsa_p5 = df_dsa.groupby('Segmento')['Valor_Venda'].sum().reset_index().sort_values(by = 'Valor_Venda',
                                                                                     ascending = False)

df_dsa_p5.head()
```

```python
# Converter os dados em valor absoluto
def autopct_format(values):
    def my_format(pct):
        total = sum(values) 
        val = int(round(pct * total / 100.0))     # calcula a porcentagem, ou seja, valor absoluto do total
        return ' $ {v:d}'.format(v = val)   # inclui o $ na saida
    return my_format
```

```python
# Plot
plt.figure(figsize = (16,6))

#autopct = autopercentual é a função feita antes, com valores absolutos
plt.pie(df_dsa_p5['Valor_Venda'],
        labels = df_dsa_p5['Segmento'],
        autopct = autopct_format(df_dsa_p5['Valor_Venda']),  
        startangle = 90)

# Limpa o círculo central
centre_circle = plt.Circle((0,0), 0.82, fc = 'white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Labels e anotações
plt.annotate(text='Total de Vendas: ' + '$ ' + str(int(sum(df_dsa_p5['Valor_Venda']))), xy = (-0.25, 0))
plt.title('Total de Vendas Por Segmento')
plt.show()
```

## Pergunta de Negócio 7 (Desafio Nível Júnior):

Os gestores da empresa estão considerando conceder diferentes faixas de descontos e gostariam de fazer uma simulação com base na regra abaixo:

- Se o Valor_Venda for maior que 1000 recebe 15% de desconto.
- Se o Valor_Venda for menor que 1000 recebe 10% de desconto.

### Quantas Vendas Receberiam 15% de Desconto?

```python
query7 = df_dsa[df_dsa['Valor_Venda'] > 1000].count()
print('Quantas vendas receberiam 15% de desconto? ',query7['Valor_Venda'])
```

```python
## Solução do professor
# Cria coluna com regra acima
df_dsa['Desconto'] = np.where(df_dsa['Valor_Venda'] > 1000, 0.15, 0.10)
```

```python
df_dsa.head()
```

```python
# Total por cada valor da variåvel
df_dsa['Desconto'].value_counts()
```

```python
print('No total 457 Vendas receberian desconto de 15%.')
```

## Pergunta de Negócio 8 (Desafio Nível Master):

### Considere Que a Empresa Decida Conceder o Desconto de 15% do Item Anterior. Qual Seria a Média do Valor de Venda Antes e Depois do Desconto?

```python
# Calculo do Valor_Venda_Desconto , usando a coluna Desconto
df_dsa['Valor_Venda_Desconto'] = df_dsa['Valor_Venda'] - (df_dsa['Valor_Venda'] * df_dsa['Desconto'])

# Média valor venda antes 15% do desconto
query8_antes = df_dsa[df_dsa['Desconto'] == 0.15]['Valor_Venda'].mean()
print('Média do valor de venda ANTES do desconto de 15%: ${:.4f}'.format(query8_antes))
print('\n')

# Média valor venda depois 15% do desconto
query8_depois = df_dsa[df_dsa['Desconto'] == 0.15]['Valor_Venda_Desconto'].mean()
print('Média do valor de venda ANTES do desconto de 15%: ${:.4f}'.format(query8_depois))
```

```python
## Solução do professor

# Filtrando as vendas antes do desconto de 15%
# loc -> realiza um slice!
df_dsa_p8_vendas_antes_desconto = df_dsa.loc[df_dsa['Desconto'] == 0.15, 'Valor_Venda']
```

```python
# Filtrando as vendas depois do desconto de 15%
df_dsa_p8_vendas_depois_desconto = df_dsa.loc[df_dsa['Desconto'] == 0.15, 'Valor_Venda_Desconto']
```

```python
# Calcula média das vendas antes do desconto 15%
media_vendas_antes_desconto = df_dsa_p8_vendas_antes_desconto.mean()
```

```python
# Calcula média das vendas depois do desconto 15%
media_vendas_depois_desconto = df_dsa_p8_vendas_depois_desconto.mean()
```

```python
print("Média das vendas antes do desconto de 15%:", round(media_vendas_antes_desconto,2))
```

```python
print("Média das vendas depois do desconto de 15%:", round(media_vendas_depois_desconto,2))
```

## Pergunta de Negócio 9 (Desafio Nível Master Ninja):

### Qual o Média de Vendas Por Segmento, Por Ano e Por Mês?

Demonstre o resultado através de gráfico de linha.

```python
# Criar coluna mês / ano:
df_dsa['Mês_Ano'] = pd.to_datetime(pd.DataFrame({'day' :1,
                                                 'month': df_dsa['Data_Pedido'].dt.month,
                                                 'year': df_dsa['Data_Pedido'].dt.year
                                                },
                                                index=df_dsa.index),format='%m-%Y',dayfirst=True)
query9 = df_dsa.groupby(['Mês_Ano', 'Segmento'], as_index = False)['Valor_Venda'] \
                        .mean() \
                        .sort_values(by=['Mês_Ano'],ascending=True) 
```

```python
plt.figure(figsize = (20,6))  

# Vendas X Segmento
for idx, segmento in enumerate(pd.unique(query9['Segmento'])):
    query9_segmento = query9[query9['Segmento'] == segmento]
    x = query9_segmento['Mês_Ano'].values
    y = query9_segmento['Valor_Venda'].values
    plt.plot(x, y, label = segmento)

plt.xlabel('Mês/Ano')
plt.ylabel('Valor médio da venda')
plt.xticks(rotation = 80)
plt.legend()
plt.show()
```

```python
## Solução do Professor
#Extraímos o mês e gravamos em uma nova variável
df_dsa['Mes'] = df_dsa['Data_Pedido'].dt.month
```

```python
# Agrupampos por ano, mês e segmento e calculamos estatísticas de agregação (agg faz +1 função de agregação,listas como funções numpy)
df_dsa_p9 = df_dsa.groupby(['Ano','Mes','Segmento'])['Valor_Venda'].agg([np.sum, np.mean, np.median])
```

```python
df_dsa_p9
```

```python
# Extrair os níveis
anos = df_dsa_p9.index.get_level_values(0)
meses = df_dsa_p9.index.get_level_values(1)
segmentos = df_dsa_p9.index.get_level_values(2)
```

```python
# Plot
plt.figure(figsize = (12,6))
sns.set()  # resetar a paleta, para a padrão
# Gráfico de relação entre 2 variáveis
fig1 = sns.relplot(kind = 'line',
                 data = df_dsa_p9,
                 y = 'mean',
                 x = meses,
                 hue = segmentos,
                 col = anos,
                 col_wrap = 4)
plt.show()
```

## Pergunta de Negócio 10 (Desafio Nível Master Ninja das Galáxias):

### Qual o Total de Vendas Por Categoria e SubCategoria, Considerando Somente as Top 12 SubCategorias? 

Demonstre tudo através de um único gráfico.

```python
## Solução do Professor

# Agrupar Categoria e Subcategoria
# Fazer soma de todos os valores numéricos do dataframe (inclui Valor_Venda)
df_dsa_p10 = df_dsa.groupby(['Categoria','SubCategoria']).sum(numeric_only = True).sort_values('Valor_Venda',
                                                                                              ascending = False).head(12)
```

```python
#Converter Valor_Venda em inteiro e classificar por categoria
# Importante ordenar por categoria, assim as subcategorias permanecem unidas
df_dsa_p10 = df_dsa_p10[['Valor_Venda']].astype(int).sort_values(by = 'Categoria').reset_index()
```

```python
df_dsa_p10
```

```python
# Dataframe somente com os totais por categoria
# Vai compor o gráfico mais externo
df_dsa_p10_cat = df_dsa_p10.groupby('Categoria').sum(numeric_only = True).reset_index()
```

```python
df_dsa_p10_cat
```

```python
# Lista de cores para categorias
cores_categorias = ['#5d00de',
                   '#03384f',
                   '#e80e27']
```

```python
# Lista de cores para subcategorias
cores_subcategorias =['#aa8cd4',
                     '#aa8cd5',
                     '#aa8cd6',
                     '#aa8cd7',
                     '#26c957',
                     '#26c958',
                     '#26c959',
                     '#26c960',
                     '#e65e65',
                     '#e65e66',
                     '#e65e67',
                     '#e65e68']
```

```python
# Plot

# Tamanho da figura
fig , ax = plt.subplots(figsize = (18,12))

# Gráfico de categorias
p1 = ax.pie(df_dsa_p10_cat['Valor_Venda'],
           radius = 1, 
           labels = df_dsa_p10_cat['Categoria'],
           wedgeprops = dict(edgecolor = 'white'),
           colors = cores_categorias)

#wedgeprops - valor da cor da linha de separação entre os setores (vai ser branco)

# Gráfico subcategorias
p2 = ax.pie(df_dsa_p10['Valor_Venda'],
           radius = 0.9,
           labels = df_dsa_p10['SubCategoria'],
           autopct = autopct_format(df_dsa_p10['Valor_Venda']),
           colors = cores_subcategorias,
           labeldistance = 0.7,
           wedgeprops = dict(edgecolor = 'white'),
           pctdistance = 0.53,
           rotatelabels = True)

#autopct_format : função feita pelo prof em questão anterior para formatar os dados para absoluto

# Limpa o centro do círculo
centre_circle = plt.Circle((0,0), 0.6, fc = 'white')

# Labels e anotações
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.annotate(text = 'Total de Vandas: ' + '$ ' + str(int(sum(df_dsa_p10['Valor_Venda']))), xy = (-0.2,0))
plt.title('Total de Vandas por Categoria e Top 12 SubCategorias')
plt.show()
```
